Lowest price: PINS or SNAP

```python
df_max = df.groupby('name')['highest'].max().reset_index()
#df_max.head()
ax = sns.barplot(x= 'name', y= 'highest', data = df_max)
```

## Showing the highest price change hourly 

### Now we can clearly see the price change every hour for each ticker! 

```python
#df.set_index('hour', inplace=True)

for i in df['name'].unique():
    df = pd.read_csv('results.csv')
    df = df[df['name'] == i]
    plt.plot(df['hour'], df['highest'])
    plt.title(str(i))
    plt.xlabel('time (hr)')
    plt.ylabel('USD')
    plt.show()
```
